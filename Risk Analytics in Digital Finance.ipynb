```python
#
# Case Study 1: Fraud Detection and Prevention (Code Corrected for Catalog Table)
#
# This version reads the data directly from the Unity Catalog table, not from a file path.
#

# --- 1. Import Necessary Libraries ---
from pyspark.sql.functions import col, hour, dayofweek
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.ml.classification import RandomForestClassifier
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

print("Libraries imported successfully.")

# --- 2. Load the Data from the Catalog Table ---
# The previous error occurred because the data is a table in the Catalog, not a file in DBFS.
# We will now load it using its proper three-part name: `catalog.schema.table`.
try:
    # Use spark.table() to read from the Unity Catalog.
    table_name = "workspace.default.finance_casestudies"
    df = spark.table(table_name)
    
    # Ensure the 'Transaction_Time' column is in the correct timestamp format.
    df = df.withColumn("Transaction_Time", col("Transaction_Time").cast("timestamp"))
    
    print(f"Successfully loaded data from table: '{table_name}'")
except Exception as e:
    print(f"ERROR: Could not read the table '{table_name}'.")
    print("Please ensure the Catalog is named 'workspace' and the Schema is 'default' in the Catalog Explorer.")
    print(f"Details: {e}")
    dbutils.notebook.exit("Table loading failed.")


# --- 3. Feature Engineering and Selection ---
# Create a new DataFrame with only the columns needed for fraud detection.
fraud_df = df.select(
    'Age', 'Gender', 'Income', 'Occupation', 'City', 'Credit_Score',
    'Amount', 'Merchant_Type', 'Transaction_Time', 'Is_Fraud'
)

# Extract hour of the day and day of the week from the timestamp.
fraud_df = fraud_df.withColumn("Transaction_Hour", hour(col("Transaction_Time")))
fraud_df = fraud_df.withColumn("Transaction_Day", dayofweek(col("Transaction_Time")))

# Rename the target column to 'label' and drop any rows with null values.
final_df = fraud_df.withColumnRenamed("Is_Fraud", "label").na.drop()
print("Feature engineering complete.")


# --- 4. Define the Machine Learning Pipeline ---
# A pipeline chains multiple data transformation and modeling steps together.

# Identify categorical (text) and numerical columns.
categorical_cols = ['Gender', 'Occupation', 'City', 'Merchant_Type']
numerical_cols = ['Age', 'Income', 'Credit_Score', 'Amount', 'Transaction_Hour', 'Transaction_Day']

stages = []

# Stages to convert Categorical Columns to a numerical format for the model.
for c in categorical_cols:
    indexer = StringIndexer(inputCol=c, outputCol=c + "_index", handleInvalid="keep")
    encoder = OneHotEncoder(inputCol=indexer.getOutputCol(), outputCol=c + "_vec")
    stages += [indexer, encoder]

# Stage to assemble all features into a single vector column named "features".
feature_cols = [c + "_vec" for c in categorical_cols] + numerical_cols
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
stages += [assembler]

# Stage for the Machine Learning Model (Random Forest).
rf = RandomForestClassifier(labelCol="label", featuresCol="features", numTrees=100, seed=42)
stages += [rf]

# Combine all stages into a single pipeline object.
pipeline = Pipeline(stages=stages)
print("ML Pipeline defined successfully.")


# --- 5. Split Data, Train, and Evaluate the Model ---
# Split the data into a training set (80%) and a testing set (20%).
(train_data, test_data) = final_df.randomSplit([0.8, 0.2], seed=42)
print(f"\nData split into {train_data.count()} training rows and {test_data.count()} testing rows.")

# Train the entire pipeline on the training data.
print("Training the fraud detection model...")
model = pipeline.fit(train_data)
print("Model training complete!")

# Make predictions on the unseen test data.
predictions = model.transform(test_data)

# --- 6. Display and Interpret Results ---
print("\n--- Model Performance Evaluation ---")

# Use a MulticlassClassificationEvaluator to measure performance.
evaluator = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction")

# Calculate Accuracy.
accuracy = evaluator.setMetricName("accuracy").evaluate(predictions)
print(f"Accuracy: {accuracy:.2%}")

# Calculate F1 Score.
f1_score = evaluator.setMetricName("f1").evaluate(predictions)
print(f"F1 Score: {f1_score:.2%}")

# Display a Confusion Matrix.
print("\nConfusion Matrix (Actual vs. Predicted):")
display(predictions.groupBy("label", "prediction").count())

# Display a sample of predictions.
print("\nSample of Predictions:")
display(predictions.select("prediction", "label", "features").limit(10))
```

```python
#
# Case Study 2: Credit Risk Scoring and Loan Default Prediction
#
# This script builds a model to predict the probability of a customer defaulting on a loan.
#

# --- 1. Import Necessary Libraries ---
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
from pyspark.sql.functions import col

print("Libraries imported successfully.")

# --- 2. Load Data from the Catalog Table ---
# This reads the data directly from the table you created in the Databricks Catalog.
try:
    table_name = "workspace.default.finance_casestudies"
    df = spark.table(table_name)
    print(f"Successfully loaded data from table: '{table_name}'")
except Exception as e:
    print(f"ERROR: Could not read the table '{table_name}'. Please ensure it exists in your catalog.")
    print(f"Details: {e}")
    dbutils.notebook.exit("Table loading failed.")

# --- 3. Feature Engineering and Selection ---
# Select columns relevant for predicting loan defaults.
# We also drop any rows with missing values in these key columns to ensure model stability.
loan_df = df.select(
    'Age', 'Gender', 'Income', 'Occupation', 'Credit_Score',
    'Loan_Amount', 'Tenure', 'EMI', 'Default_Status'
).na.drop()

# Rename the target column to 'label', which is the standard name for PySpark ML.
final_df = loan_df.withColumnRenamed("Default_Status", "label")
print("Feature selection and preparation complete.")


# --- 4. Define the Machine Learning Pipeline ---
# The pipeline will process the data and train the model in a series of stages.

# Identify categorical (text) and numerical columns.
categorical_cols = ['Gender', 'Occupation']
numerical_cols = ['Age', 'Income', 'Credit_Score', 'Loan_Amount', 'Tenure', 'EMI']

stages = []

# Stages to convert categorical string columns into a numerical vector format.
for c in categorical_cols:
    indexer = StringIndexer(inputCol=c, outputCol=c + "_index", handleInvalid="keep")
    encoder = OneHotEncoder(inputCol=indexer.getOutputCol(), outputCol=c + "_vec")
    stages += [indexer, encoder]

# Stage to assemble all processed features into a single vector.
feature_cols = [c + "_vec" for c in categorical_cols] + numerical_cols
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
stages += [assembler]

# Stage for the Logistic Regression model. We set elasticNetParam=0.8 for a good balance (like Elastic Net).
lr = LogisticRegression(labelCol="label", featuresCol="features", elasticNetParam=0.8)
stages += [lr]

# Combine all stages into a single pipeline.
pipeline = Pipeline(stages=stages)
print("ML Pipeline for credit risk defined.")


# --- 5. Split Data, Train, and Evaluate the Model ---
# Split the data into training (80%) and testing (20%) sets.
(train_data, test_data) = final_df.randomSplit([0.8, 0.2], seed=42)
print(f"\nData split into {train_data.count()} training rows and {test_data.count()} testing rows.")

# Train the pipeline on the training data.
print("Training the loan default model...")
model = pipeline.fit(train_data)
print("Model training complete!")

# Make predictions on the test data.
predictions = model.transform(test_data)

# --- 6. Display Model Performance ---
print("\n--- Model Performance Evaluation ---")

# Evaluator for metrics like Accuracy and F1 Score.
evaluator_multi = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction")
# Evaluator for Area Under ROC (AUC), a key metric for risk models.
evaluator_auc = BinaryClassificationEvaluator(labelCol="label", rawPredictionCol="rawPrediction", metricName="areaUnderROC")

accuracy = evaluator_multi.setMetricName("accuracy").evaluate(predictions)
f1_score = evaluator_multi.setMetricName("f1").evaluate(predictions)
auc = evaluator_auc.evaluate(predictions)

print(f"Accuracy: {accuracy:.2%}")
print(f"F1 Score: {f1_score:.2%}")
print(f"Area Under ROC (AUC): {auc:.2%}") # An AUC > 70% is generally considered acceptable.

# Display the raw predictions DataFrame which will be used for visualization.
print("\nModel predictions are ready for visualization.")
display(predictions)
```

```python
#
# Case Study 3: Customer Segmentation (Corrected Code)
#
# This script uses K-Means clustering to group customers into distinct segments.
# The 'import pandas as pd' line has been added to fix the previous error.
#

# --- 1. Import Necessary Libraries ---
from pyspark.ml.feature import VectorAssembler, StandardScaler
from pyspark.ml.clustering import KMeans
from pyspark.sql.functions import col
from pyspark.ml import Pipeline # Added Pipeline to the import list for clarity
import matplotlib.pyplot as plt
import pandas as pd # <--- THIS LINE FIXES THE ERROR

print("Libraries imported successfully.")

# --- 2. Load Data from the Catalog Table ---
try:
    table_name = "workspace.default.finance_casestudies"
    df = spark.table(table_name)
    print(f"Successfully loaded data from table: '{table_name}'")
except Exception as e:
    print(f"ERROR: Could not read the table '{table_name}'.")
    dbutils.notebook.exit("Table loading failed.")

# --- 3. Feature Selection and Preparation ---
feature_cols = ['Age', 'Income', 'Credit_Score', 'Loan_Amount', 'Investment_Amount']
segment_df = df.select(feature_cols).na.drop()

assembler = VectorAssembler(inputCols=feature_cols, outputCol="features_unscaled")
scaler = StandardScaler(inputCol="features_unscaled", outputCol="features", withStd=True, withMean=False)

preprocessing_pipeline = Pipeline(stages=[assembler, scaler])
prepped_df = preprocessing_pipeline.fit(segment_df).transform(segment_df)

print("Feature selection and scaling complete.")

# --- 4. Find the Optimal Number of Clusters (Elbow Method) ---
print("\nCalculating model cost (WSSSE) for different values of k...")
costs = []
k_values = range(2, 10)

for k in k_values:
    kmeans = KMeans(k=k, seed=42, featuresCol='features')
    model = kmeans.fit(prepped_df)
    cost = model.summary.trainingCost
    costs.append(cost)
    print(f"For k={k}, Within Set Sum of Squared Errors (WSSSE) = {cost:.2f}")

# Plot the Elbow Method graph.
elbow_plot_df = pd.DataFrame({'k': list(k_values), 'cost': costs}) # This line now works correctly
plt.figure(figsize=(10, 6))
plt.plot(elbow_plot_df['k'], elbow_plot_df['cost'], marker='o', linestyle='--')
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Cost (WSSSE)")
plt.title("Elbow Method for Optimal k")
plt.grid(True)
print("\nDisplaying Elbow Plot. Look for the 'elbow' where the cost reduction slows down.")
display(plt.show())

# --- 5. Train the Final K-Means Model ---
OPTIMAL_K = 4
kmeans = KMeans(k=OPTIMAL_K, seed=42, featuresCol='features', predictionCol='segment')

print(f"\nTraining final K-Means model with k={OPTIMAL_K} clusters...")
final_model = kmeans.fit(prepped_df)
print("Model training complete!")

predictions = final_model.transform(prepped_df)

# --- 6. Display Results for Visualization ---
print(f"\nCustomer segmentation complete. The 'segment' column shows the cluster for each customer.")
# This final DataFrame is what you will use for the dashboard visuals.
display(predictions.select(feature_cols + ['segment']))
```

```python
#
# Case Study 4: Stock Market Sentiment Analysis
#
# This script builds a Linear Regression model to see if stock sentiment
# can predict stock returns.
#

# --- 1. Import Necessary Libraries ---
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.regression import LinearRegression
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.sql.functions import col, corr

print("Libraries imported successfully.")

# --- 2. Load Data from the Catalog Table ---
try:
    table_name = "workspace.default.finance_casestudies"
    df = spark.table(table_name)
    print(f"Successfully loaded data from table: '{table_name}'")
except Exception as e:
    print(f"ERROR: Could not read the table '{table_name}'.")
    dbutils.notebook.exit("Table loading failed.")

# --- 3. Feature Selection and Initial Analysis ---
# Select only the columns relevant for this analysis.
# Drop any rows where either sentiment or return is missing.
stock_df = df.select('Stock_Sentiment_Score', 'Stock_Return').na.drop()

# Before building a model, let's calculate the correlation between the two variables.
# This tells us the strength and direction of their linear relationship.
correlation = stock_df.select(corr('Stock_Sentiment_Score', 'Stock_Return')).collect()[0][0]
print(f"\nCorrelation between Sentiment Score and Stock Return: {correlation:.4f}")
# A correlation near 0 suggests a weak linear relationship.

# --- 4. Prepare Data for Machine Learning ---
# The model requires the feature(s) to be in a single vector column.
assembler = VectorAssembler(inputCols=['Stock_Sentiment_Score'], outputCol='features')

# Rename the target column ('Stock_Return') to 'label' for PySpark ML.
final_df = stock_df.withColumnRenamed('Stock_Return', 'label')
final_df = assembler.transform(final_df)

print("Data has been prepared for the regression model.")

# --- 5. Split Data and Train the Model ---
# Split data into training (80%) and testing (20%) sets.
(train_data, test_data) = final_df.randomSplit([0.8, 0.2], seed=42)
print(f"\nData split into {train_data.count()} training rows and {test_data.count()} testing rows.")

# Define the Linear Regression model.
lr = LinearRegression(featuresCol='features', labelCol='label')

# Train the model on the training data.
print("Training the Linear Regression model...")
lr_model = lr.fit(train_data)
print("Model training complete!")

# --- 6. Analyze and Evaluate the Model ---
# Print the model's learned formula: return = coefficient * sentiment + intercept
print(f"\nModel Coefficients: {lr_model.coefficients[0]:.4f}")
print(f"Model Intercept: {lr_model.intercept:.4f}")

# Make predictions on the test data.
predictions = lr_model.transform(test_data)

# Evaluate the model's performance.
evaluator = RegressionEvaluator(labelCol="label", predictionCol="prediction")

# RMSE (Root Mean Squared Error): The average error of the model's predictions. Lower is better.
rmse = evaluator.setMetricName("rmse").evaluate(predictions)
# R-squared (R2): The proportion of variance in the return that is predictable from the sentiment. Closer to 1 is better.
r2 = evaluator.setMetricName("r2").evaluate(predictions)

print("\n--- Model Performance Evaluation ---")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R2): {r2:.4f}")


# --- 7. Display Results for Visualization ---
# The final 'predictions' DataFrame contains everything needed for the dashboard.
print("\nPredictions are ready for visualization.")
display(predictions.select(col("Stock_Sentiment_Score"), col("label").alias("Actual_Return"), col("prediction").alias("Predicted_Return")))
```

```python
#
# Case Study 5: Insurance Claim Analytics and Risk Prediction
#
# This script builds a Decision Tree model to predict the likelihood of an insurance claim.
#

# --- 1. Import Necessary Libraries ---
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, OneHotEncoder, VectorAssembler
from pyspark.ml.classification import DecisionTreeClassifier
from pyspark.ml.evaluation import BinaryClassificationEvaluator, MulticlassClassificationEvaluator
from pyspark.sql.functions import col
import pandas as pd

print("Libraries imported successfully.")

# --- 2. Load Data from the Catalog Table ---
try:
    table_name = "workspace.default.finance_casestudies"
    df = spark.table(table_name)
    print(f"Successfully loaded data from table: '{table_name}'")
except Exception as e:
    print(f"ERROR: Could not read the table '{table_name}'.")
    dbutils.notebook.exit("Table loading failed.")

# --- 3. Feature Selection and Preparation ---
# Select features that are likely to influence whether a claim is filed.
# Drop rows with missing values in these key columns.
claim_df = df.select(
    'Age', 'Gender', 'Income', 'Occupation', 'Credit_Score',
    'Risk_Score', 'Claim_Status'
).na.drop()

# Rename the target column to 'label' for PySpark ML.
final_df = claim_df.withColumnRenamed("Claim_Status", "label")
print("Feature selection and preparation complete.")

# --- 4. Define the Machine Learning Pipeline ---
# The pipeline will process the data and train the model in a series of stages.

# Identify categorical and numerical columns.
categorical_cols = ['Gender', 'Occupation']
numerical_cols = ['Age', 'Income', 'Credit_Score', 'Risk_Score']

stages = []

# Stages to convert categorical string columns into a numerical vector format.
for c in categorical_cols:
    indexer = StringIndexer(inputCol=c, outputCol=c + "_index", handleInvalid="keep")
    encoder = OneHotEncoder(inputCol=indexer.getOutputCol(), outputCol=c + "_vec")
    stages += [indexer, encoder]

# Stage to assemble all processed features into a single vector.
feature_cols = [c + "_vec" for c in categorical_cols] + numerical_cols
assembler = VectorAssembler(inputCols=feature_cols, outputCol="features")
stages += [assembler]

# Stage for the Decision Tree model.
dt = DecisionTreeClassifier(labelCol="label", featuresCol="features", seed=42)
stages += [dt]

# Combine all stages into a single pipeline.
pipeline = Pipeline(stages=stages)
print("ML Pipeline for claim prediction defined.")

# --- 5. Split Data, Train, and Evaluate the Model ---
(train_data, test_data) = final_df.randomSplit([0.8, 0.2], seed=42)
print(f"\nData split into {train_data.count()} training rows and {test_data.count()} testing rows.")

print("Training the insurance claim prediction model...")
model = pipeline.fit(train_data)
print("Model training complete!")

# Make predictions on the test data.
predictions = model.transform(test_data)

# --- 6. Display Model Performance ---
print("\n--- Model Performance Evaluation ---")

evaluator_multi = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction")
evaluator_auc = BinaryClassificationEvaluator(labelCol="label", rawPredictionCol="rawPrediction", metricName="areaUnderROC")

accuracy = evaluator_multi.setMetricName("accuracy").evaluate(predictions)
f1_score = evaluator_multi.setMetricName("f1").evaluate(predictions)
auc = evaluator_auc.evaluate(predictions)

print(f"Accuracy: {accuracy:.2%}")
print(f"F1 Score: {f1_score:.2%}")
print(f"Area Under ROC (AUC): {auc:.2%}")

# Display the predictions DataFrame for visualization.
print("\nModel predictions are ready for visualization.")
display(predictions)
```

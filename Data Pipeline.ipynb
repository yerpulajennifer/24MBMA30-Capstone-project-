```python
# ==============================================================================
# PART 0: SETUP AND CONFIGURATION
# ==============================================================================

# --- Import All Necessary Libraries ---
from pyspark.sql.functions import col, hour, dayofweek, corr
from pyspark.ml import Pipeline
from pyspark.ml.feature import StringIndexer, VectorAssembler, StandardScaler
from pyspark.ml.classification import LogisticRegression, DecisionTreeClassifier
from pyspark.ml.regression import LinearRegression
from pyspark.ml.clustering import KMeans
from pyspark.ml.evaluation import MulticlassClassificationEvaluator, BinaryClassificationEvaluator, RegressionEvaluator
import matplotlib.pyplot as plt
import pandas as pd

print("All necessary libraries have been imported successfully.")

# --- Load the Dataset from the Databricks Catalog ---
try:
    table_name = "workspace.default.finance_casestudies"
    df_base = spark.table(table_name)
    df_base = df_base.withColumn("Transaction_Time", col("Transaction_Time").cast("timestamp"))
    print(f"Successfully loaded and prepared the base DataFrame from table: '{table_name}'")
    print(f"Total records in the dataset: {df_base.count()}")
except Exception as e:
    print(f"FATAL ERROR: Could not read the table '{table_name}'.")
    dbutils.notebook.exit("Data loading failed. Halting execution.")

# ==============================================================================
# CASE STUDY 1: FRAUD DETECTION AND PREVENTION (FINAL FIX)
# ==============================================================================

print("\n" + "="*80)
print("STARTING CASE STUDY 1: FRAUD DETECTION AND PREVENTION")
print("="*80)

# --- 1a. Feature Engineering & Simplified Selection ---
# To prevent memory errors, we are removing high-cardinality text features like City, Occupation, and Merchant_Type.
fraud_df = df_base.select(
    'Age', 'Gender', 'Income', 'Credit_Score', 'Amount', 'Transaction_Time', 'Is_Fraud'
).na.drop()
fraud_df = fraud_df.withColumn("Transaction_Hour", hour(col("Transaction_Time")))
fraud_df = fraud_df.withColumn("Transaction_Day", dayofweek(col("Transaction_Time")))
final_fraud_df = fraud_df.withColumnRenamed("Is_Fraud", "label")

# --- 1b. Define a Lightweight ML Pipeline ---
categorical_cols_1 = ['Gender'] # Using only low-cardinality category
numerical_cols_1 = ['Age', 'Income', 'Credit_Score', 'Amount', 'Transaction_Hour', 'Transaction_Day']
stages_1 = []
indexer_1 = StringIndexer(inputCol='Gender', outputCol='Gender_idx', handleInvalid="keep")
stages_1.append(indexer_1)

assembler_inputs_1 = ['Gender_idx'] + numerical_cols_1
assembler_1 = VectorAssembler(inputCols=assembler_inputs_1, outputCol="features")
lr_fraud_1 = LogisticRegression(labelCol="label", featuresCol="features")
stages_1 += [assembler_1, lr_fraud_1]
pipeline_1 = Pipeline(stages=stages_1)

# --- 1c. Train and Evaluate Model ---
(train_1, test_1) = final_fraud_df.randomSplit([0.8, 0.2], seed=42)
print(f"Training Fraud Detection model on {train_1.count()} records...")
fraud_model = pipeline_1.fit(train_1)
fraud_predictions = fraud_model.transform(test_1)
evaluator_1 = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction")
accuracy_1 = evaluator_1.setMetricName("accuracy").evaluate(fraud_predictions)
f1_1 = evaluator_1.setMetricName("f1").evaluate(fraud_predictions)

print("\n--- Case 1: Fraud Model Evaluation ---")
print(f"Accuracy: {accuracy_1:.2%}")
print(f"F1 Score: {f1_1:.2%}")

# --- 1d. Display Final Output for Visualization ---
print("\nCASE 1 FINAL OUTPUT: This table is ready for visualization.")
display(fraud_predictions)

# ==============================================================================
# CASE STUDY 2: CREDIT RISK AND LOAN DEFAULT PREDICTION (MODIFIED)
# ==============================================================================

print("\n" + "="*80)
print("STARTING CASE STUDY 2: CREDIT RISK AND LOAN DEFAULT PREDICTION")
print("="*80)

# --- 2a. Feature Engineering & Simplified Selection ---
# Removing 'Occupation' to ensure a small model size.
loan_df = df_base.select(
    'Age', 'Gender', 'Income', 'Credit_Score', 'Loan_Amount', 'Tenure', 'EMI', 'Default_Status'
).na.drop()
final_loan_df = loan_df.withColumnRenamed("Default_Status", "label")

# --- 2b. Define a Lightweight ML Pipeline ---
categorical_cols_2 = ['Gender']
numerical_cols_2 = ['Age', 'Income', 'Credit_Score', 'Loan_Amount', 'Tenure', 'EMI']
stages_2 = []
indexer_2 = StringIndexer(inputCol='Gender', outputCol='Gender_idx', handleInvalid="keep")
stages_2.append(indexer_2)

assembler_inputs_2 = ['Gender_idx'] + numerical_cols_2
assembler_2 = VectorAssembler(inputCols=assembler_inputs_2, outputCol="features")
lr_2 = LogisticRegression(labelCol="label", featuresCol="features")
stages_2 += [assembler_2, lr_2]
pipeline_2 = Pipeline(stages=stages_2)

# --- 2c. Train and Evaluate Model ---
(train_2, test_2) = final_loan_df.randomSplit([0.8, 0.2], seed=42)
print(f"Training Loan Default model on {train_2.count()} records...")
loan_model = pipeline_2.fit(train_2)
loan_predictions = loan_model.transform(test_2)
evaluator_multi_2 = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction")
evaluator_auc_2 = BinaryClassificationEvaluator(labelCol="label", rawPredictionCol="rawPrediction")
accuracy_2 = evaluator_multi_2.setMetricName("accuracy").evaluate(loan_predictions)
f1_2 = evaluator_multi_2.setMetricName("f1").evaluate(loan_predictions)
auc_2 = evaluator_auc_2.setMetricName("areaUnderROC").evaluate(loan_predictions)

print("\n--- Case 2: Loan Default Model Evaluation ---")
print(f"Accuracy: {accuracy_2:.2%}")
print(f"F1 Score: {f1_2:.2%}")
print(f"Area Under ROC (AUC): {auc_2:.2%}")

# --- 2d. Display Final Output for Visualization ---
print("\nCASE 2 FINAL OUTPUT: This table is ready for visualization.")
display(loan_predictions)

# ==============================================================================
# CASE STUDY 3: CUSTOMER SEGMENTATION
# ==============================================================================

print("\n" + "="*80)
print("STARTING CASE STUDY 3: CUSTOMER SEGMENTATION")
print("="*80)

# --- 3a. Feature Selection and Preparation ---
feature_cols_3 = ['Age', 'Income', 'Credit_Score', 'Loan_Amount', 'Investment_Amount']
segment_df = df_base.select(feature_cols_3).na.drop()
assembler_3 = VectorAssembler(inputCols=feature_cols_3, outputCol="features_unscaled")
scaler_3 = StandardScaler(inputCol="features_unscaled", outputCol="features", withStd=True, withMean=False)
preprocessing_pipeline_3 = Pipeline(stages=[assembler_3, scaler_3])
prepped_df_3 = preprocessing_pipeline_3.fit(segment_df).transform(segment_df)

# --- 3b. Find Optimal Clusters (Elbow Method) ---
print("\nCalculating model cost for Elbow Method...")
costs = []
k_values = range(2, 10)
for k in k_values:
    kmeans = KMeans(k=k, seed=42, featuresCol='features')
    model = kmeans.fit(prepped_df_3)
    costs.append(model.summary.trainingCost)
elbow_plot_df = pd.DataFrame({'k': list(k_values), 'cost': costs})
plt.figure(figsize=(10, 6))
plt.plot(elbow_plot_df['k'], elbow_plot_df['cost'], marker='o', linestyle='--')
plt.xlabel("Number of Clusters (k)"); plt.ylabel("Cost (WSSSE)"); plt.title("Elbow Method")
plt.grid(True)
print("Displaying Elbow Plot...")
display(plt.show())

# --- 3c. Train Final K-Means Model ---
OPTIMAL_K = 4
kmeans = KMeans(k=OPTIMAL_K, seed=42, featuresCol='features', predictionCol='segment')
print(f"\nTraining final K-Means model with k={OPTIMAL_K} clusters...")
segment_model = kmeans.fit(prepped_df_3)
segment_predictions = segment_model.transform(prepped_df_3)

# --- 3d. Display Final Output for Visualization ---
print("\nCASE 3 FINAL OUTPUT: This table is ready for visualization.")
display(segment_predictions.select(feature_cols_3 + ['segment']))

# ==============================================================================
# CASE STUDY 4: STOCK MARKET SENTIMENT ANALYSIS
# ==============================================================================

print("\n" + "="*80)
print("STARTING CASE STUDY 4: STOCK MARKET SENTIMENT ANALYSIS")
print("="*80)

# --- 4a. Feature Selection and Analysis ---
stock_df = df_base.select('Stock_Sentiment_Score', 'Stock_Return').na.drop()
correlation = stock_df.select(corr('Stock_Sentiment_Score', 'Stock_Return')).first()[0]
print(f"Correlation between Sentiment Score and Stock Return: {correlation:.4f}")

# --- 4b. Prepare Data and Train Model ---
assembler_4 = VectorAssembler(inputCols=['Stock_Sentiment_Score'], outputCol='features')
final_stock_df = stock_df.withColumnRenamed('Stock_Return', 'label')
final_stock_df = assembler_4.transform(final_stock_df)
(train_4, test_4) = final_stock_df.randomSplit([0.8, 0.2], seed=42)
print(f"Training Linear Regression model on {train_4.count()} records...")
lr_4 = LinearRegression(featuresCol='features', labelCol='label')
sentiment_model = lr_4.fit(train_4)
sentiment_predictions = sentiment_model.transform(test_4)

# --- 4c. Evaluate Model ---
evaluator_4 = RegressionEvaluator(labelCol="label", predictionCol="prediction")
rmse_4 = evaluator_4.setMetricName("rmse").evaluate(sentiment_predictions)
r2_4 = evaluator_4.setMetricName("r2").evaluate(sentiment_predictions)
print("\n--- Case 4: Sentiment Model Evaluation ---")
print(f"Root Mean Squared Error (RMSE): {rmse_4:.4f}")
print(f"R-squared (R2): {r2_4:.4f}")

# --- 4d. Display Final Output for Visualization ---
print("\nCASE 4 FINAL OUTPUT: This table is ready for visualization.")
display(sentiment_predictions.select(col("Stock_Sentiment_Score"), col("label").alias("Actual_Return"), col("prediction").alias("Predicted_Return")))

# ==============================================================================
# CASE STUDY 5: INSURANCE CLAIM ANALYTICS AND RISK PREDICTION (MODIFIED)
# ==============================================================================

print("\n" + "="*80)
print("STARTING CASE STUDY 5: INSURANCE CLAIM ANALYTICS")
print("="*80)

# --- 5a. Feature Engineering & Simplified Selection ---
# Removing 'Occupation' to ensure a small model size.
claim_df = df_base.select(
    'Age', 'Gender', 'Income', 'Credit_Score', 'Risk_Score', 'Claim_Status'
).na.drop()
final_claim_df = claim_df.withColumnRenamed("Claim_Status", "label")

# --- 5b. Define a Lightweight ML Pipeline ---
categorical_cols_5 = ['Gender']
numerical_cols_5 = ['Age', 'Income', 'Credit_Score', 'Risk_Score']
stages_5 = []
indexer_5 = StringIndexer(inputCol='Gender', outputCol='Gender_idx', handleInvalid="keep")
stages_5.append(indexer_5)
assembler_inputs_5 = ['Gender_idx'] + numerical_cols_5
assembler_5 = VectorAssembler(inputCols=assembler_inputs_5, outputCol="features")
dt_5 = DecisionTreeClassifier(labelCol="label", featuresCol="features", seed=42, maxDepth=5)
stages_5 += [assembler_5, dt_5]
pipeline_5 = Pipeline(stages=stages_5)

# --- 5c. Train and Evaluate Model ---
(train_5, test_5) = final_claim_df.randomSplit([0.8, 0.2], seed=42)
print(f"Training Insurance Claim model on {train_5.count()} records...")
claim_model = pipeline_5.fit(train_5)
claim_predictions = claim_model.transform(test_5)
evaluator_multi_5 = MulticlassClassificationEvaluator(labelCol="label", predictionCol="prediction")
evaluator_auc_5 = BinaryClassificationEvaluator(labelCol="label", rawPredictionCol="rawPrediction")
accuracy_5 = evaluator_multi_5.setMetricName("accuracy").evaluate(claim_predictions)
f1_5 = evaluator_multi_5.setMetricName("f1").evaluate(claim_predictions)
auc_5 = evaluator_auc_5.setMetricName("areaUnderROC").evaluate(claim_predictions)

print("\n--- Case 5: Insurance Claim Model Evaluation ---")
print(f"Accuracy: {accuracy_5:.2%}")
print(f"F1 Score: {f1_5:.2%}")
print(f"Area Under ROC (AUC): {auc_5:.2%}")

# --- 5d. Create and Display Feature Importance Table ---
dt_model_5 = claim_model.stages[-1]
feature_cols_5 = claim_model.stages[-2].getInputCols()
importance_scores_5 = dt_model_5.featureImportances.toArray().tolist()
feature_importance_list_5 = list(zip(feature_cols_5, importance_scores_5))
importance_df_5 = spark.createDataFrame(feature_importance_list_5, ["Feature", "Importance"])

print("\nCASE 5 FEATURE IMPORTANCE: This table is ready for visualization.")
display(importance_df_5.orderBy(col("Importance"), ascending=False))

print("\nCASE 5 FINAL OUTPUT: This table is ready for visualization.")
display(claim_predictions)

print("\n" + "="*80)
print("ALL 5 CASE STUDIES HAVE BEEN EXECUTED SUCCESSFULLY.")
print("="*80)
```
